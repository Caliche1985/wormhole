```python
import sympy as sp
import numpy as np
import math

from matplotlib import pyplot as plt
from scipy.optimize import fsolve
```

The line element of a spherically symmetric wormhole is given by

\begin{equation}
ds^2=-e^{2\Phi(r)}dt^2+\frac{dr^2}{\left(1-\frac{b(r)}{r}\right)}+r^2(d\theta^2+\sin^2\theta d\varphi^2)
\end{equation}

$\Phi(r)$ and $b(r)$ are the shift function and the shape function, respectively. In this manuscrit, we use $b(r)=b^2_0/r$ as the shape function, where $b_0$ is the throat of the wormhole. In general, $b_0$ is interpreted as the mass of the wormhole in the Newtonian limit (see Cosimo PRD 78, 084005). On the other hand, a commun choise for the redshift function is ([Harko PRD 78, 084005][Cosimo PRD 90, 024071])

\begin{equation}
\Phi(r)=-\frac{b_0}{r}.
\end{equation}

This chose of $\Phi(r)$ is based on the behaviour of circular orbits around wormholes, which are unstable when $\partial\Phi/\partial r <0$ (due to the
outward gravitational repulsion) and stable when $\partial\Phi/\partial r >0$[Harko PRD 78, 084005]. 

Using the proper distant $l$, related to $r$ by the relation $r^2=b^2_0+l^2$, the line element takes the following form

\begin{equation}
ds^2=-e^{2\Phi(l)}dt^2+dl^2+(b^2_0+l^2)(d\theta^2+\sin^2\theta d\phi^2),
\end{equation}

with

\begin{equation}
\Phi(l)=-\frac{b_0}{\sqrt{b^2_0+l^2}}.
\end{equation}

In this way, we are alble to consider both the lower ($l<0$) and the upper ($l>0$) Universes. The throat, when $r = b_0$, is located at $l=0$. 

Since the spacetime is static and spherically symmetric, the line element allows two killing vector fields

\begin{equation}
\begin{array}{ccc}
\xi^\alpha=\delta^\alpha_t&\text{and}&\xi^\alpha=\delta^\alpha_\phi
\end{array}
\end{equation}

These Killing vectors are related to the conserved quantities in the spacetime by the following relation 

\begin{equation}
p_\alpha\xi^\alpha - \frac{1}{2}S^{\alpha\beta}\nabla_\beta \xi_\alpha = \text{constant}.
\end{equation}

Due to the symmetry and anti-symmetry of the Christoffel symbols and the spin tensor, respectively, the second term in the last equation can be expressed as 

\begin{equation}
\begin{aligned}
S^{\mu\beta}\nabla_\beta\xi_\mu&= S^{\mu\beta}\partial_\beta\xi_\mu-\frac{1}{2}\left(S^{\mu\beta}-S^{\beta\mu}\right)\Gamma^\lambda_{\mu\beta}\xi_\lambda\\
&=S^{\mu\beta}\partial_\beta\xi_\mu-\frac{1}{2}\left(S^{\mu\beta}\Gamma^\lambda_{\mu\beta}\xi_\lambda-S^{\beta\mu}\Gamma^\lambda_{\mu\beta}\xi_\lambda\right).
\end{aligned}
\end{equation}

Changin $\beta\rightarrow\mu$ and $\mu\rightarrow\beta$ in the second term inside the brakets, we get

\begin{equation}
S^{\mu\beta}\nabla_\beta\xi_\mu=S^{\mu\beta}\partial_\beta\xi_\mu-\frac{1}{2}\left(S^{\mu\beta}\Gamma^\lambda_{\mu\beta}\xi_\lambda-S^{\mu\beta}\Gamma^\lambda_{\beta\mu}\xi_\lambda\right)=S^{\mu\beta}\partial_\beta\xi_\mu.
\end{equation}

Hence, the equation for the conserved quantities reduces to

\begin{equation}
p_\alpha\xi^\alpha - \frac{1}{2}S^{\alpha\beta}\partial_\beta\xi_\alpha = \text{constant}.
\end{equation}

Using the Killing vectors, we obtain the equations for the concerved quantities: the energy $E$ and the Angular momentum $L$. These equations are given by

\begin{equation}
\begin{aligned}
-E&=p_t-\frac{1}{2}S^{\mu\beta}\partial_\beta g_{t\mu}=p_t-\frac{1}{2}S^{tl}\partial_lg_{tt}\\
L&=p_\phi+\frac{1}{2}S^{\beta\mu}\partial_\beta g_{\phi\mu}=p_\phi+\frac{1}{2}S^{l\phi}\partial_lg_{\phi\phi}+\frac{1}{2}S^{\theta\phi}\partial_\theta g_{\phi\phi}.
\end{aligned}
\end{equation}

Usually, for astrophysical considerations, it is useful to consider the motion on the equatorial plane ($\theta =\pi/2$). In this particular case, the metric components only depend on $l$ and $p^\theta=0$. Moreover, when constrained to the equatorial plane, the components $S^{\theta\alpha}$ vanish and of the spin tensor reduces its components to three: $S^{tl}$, $S^{t\phi}$ and $S^{l\phi}$. Hence, the conserved quantities reduce to 

\begin{equation}
\begin{aligned}
E&=-p_t+S^{tl}\partial_l g_{tt}\\
L&=p_\phi+\frac{1}{2}S^{l\phi}\partial_lg_{\phi\phi}.
\end{aligned}
\end{equation}

Using the the Tulczyjew-SSC condition

\begin{equation}
S^{\alpha\beta}p_\alpha=0,
\end{equation}

it is possible to find the relation between $S^{tl}$, $S^{t\phi}$ and $S^{l\phi}$. From it we obtain

\begin{equation}
\begin{aligned}
S^{\alpha t}p_\alpha&=S^{l t}p_l+S^{\phi t}p_\phi=0\\
S^{\alpha\phi}p_\alpha&=S^{l \phi}p_l+S^{t\phi}p_t=0,
\end{aligned}
\end{equation}

from which

\begin{equation}
\begin{aligned}
S^{t\phi}&=\frac{p_l}{p_\phi}S^{lt}=-\frac{p_l}{p_\phi}S^{tl}\\
S^{l\phi}&=-\frac{p_t}{p_l}S^{t\phi}=\frac{p_t}{p_\phi}S^{tl}.
\end{aligned}
\end{equation}

On the other hand, from the normalization condition 

\begin{equation}
g^{\alpha\mu}p_\mu p_\alpha=p^\alpha p_\alpha=p^t p_t+p^l p_l+p^\phi p_\phi=-m^2,
\end{equation}

we have

\begin{equation}
(p_l)^2=\frac{1}{g^{ll}}\left(-g^{tt}(p_t)^2-g^{\phi\phi}(p_\phi)^2-m^2\right),
\end{equation}

and from the conservation of the angular momemtum 

\begin{equation}
S^2=\frac{1}{2}S^{\alpha\beta}S_{\alpha\beta},
\end{equation}

we have

\begin{equation}
S^2=\frac{(S^{tl})^2}{p^2_{\phi}}(g_{tt}g_{ll}p^2_\phi+g_{\phi\phi}g_{tt}p^2_l+g_{ll}g_{\phi\phi}p^2_t), 
\end{equation}

where we used the relation between $S^{tl}$, $S^{t\phi}$ and $S^{l\phi}$. Hence after replacing the normalization condition, we obtain 

\begin{equation}
S^{tl}=\frac{s p_\phi}{\sqrt{-g_{tt}g_{ll}g_{\phi\phi}}}.
\end{equation}

Here $s=S/m$ can be positive or negative, which represents the direction of the spin with respect to the direction of $p_\phi$. 

Using the last equation for $S^{tl}$ and $S^{l\phi}=(p_t/p_\phi)S^{tl}$ in the equations for $E$ and $L$, we obtain 

\begin{equation}
\begin{aligned}
E&=-p_t+\frac{1}{2}\frac{sg'_{tt}p_\phi}{\sqrt{-g_{tt}g_{ll}g_{\phi\phi}}}\\
L&=p_\phi+\frac{1}{2}\frac{sg'_{\phi\phi}p_t}{\sqrt{-g_{tt}g_{ll}g_{\phi\phi}}}.
\end{aligned}
\end{equation}

In the last expression ${}'$ is the partial derivate with respect to $l$. Then, solving the system for $p_t$ and $p_\phi$, we obtain

\begin{equation}
\begin{aligned}
p_t&=\frac{-E+\mathcal{A}sL}{1-\mathcal{D}s^2}\\
p_\phi&=\frac{L+\mathcal{B}sE}{1-\mathcal{D}s^2}
\end{aligned}
\end{equation}

with

\begin{equation}
\begin{aligned}
\mathcal{A}&=\frac{g'_{tt}}{2\sqrt{-g_{tt}g_{ll}g_{\phi\phi}}}\\
\mathcal{B}&=\frac{g'_{\phi\phi}}{2\sqrt{-g_{tt}g_{ll}g_{\phi\phi}}}\\
\mathcal{D}&=\frac{g'_{tt}g'_{\phi\phi}}{4g_{tt}g_{ll}g_{\phi\phi}}
\end{aligned}
\end{equation}

Then, after replacing the expressions for $p_t$ and $p_\phi$ into the equation for $(p_l)^2$, we obtain 

\begin{equation}
(p_l)^2=\frac{1}{\alpha}(\beta E^2 + 2\delta L E + \sigma L^2 - \rho ),
\end{equation}

which can be expressed as

\begin{equation}
(p_l)^2=\frac{\beta}{\alpha}(E-V_+)(E-V_-)
\end{equation}

with 

\begin{equation}
V_{\pm}=-\frac{\delta L}{\beta}\pm\sqrt{\frac{\delta^2 L^2}{\beta^2}+\frac{\rho-\sigma L^2}{\beta}}
\end{equation}

where

\begin{equation}
\begin{aligned}
\alpha&=g^{ll}\left(1-\frac{g'_{tt}g'_{\phi\phi} s^2}{4g_{tt}g_{ll}g_{\phi\phi}}\right)^2\\
%
\beta&=-g^{tt}+\frac{g^{\phi\phi}(g'_{\phi\phi})^2s^2}{4g_{tt}g_{ll}g_{\phi\phi}}\\
%
\delta&=\frac{(g^{tt}g'_{tt}-g^{\phi\phi}g'_{\phi\phi})s}{2\sqrt{-g_{tt}g_{ll}g_{\phi\phi}}}\\
%
\sigma&=-g^{\phi\phi}+\frac{g^{tt}(g'_{tt})^2s^2}{4g_{tt}g_{ll}g_{\phi\phi}}\\
%
\rho&=m^2\left(1-\frac{g'_{tt}g'_{\phi\phi}s^2}{4g_{tt}g_{ll}g_{\phi\phi}}\right)^2
\end{aligned}
\end{equation}

```python
l = sp.Symbol('l')
b = sp.Symbol('b')
s = sp.Symbol('s')
L = sp.Symbol('L')
```

# Defintions

Here we define the variables $\alpha$, $\beta$, $\delta$, $\sigma$ and $\rho$. Hence we have

\begin{equation}
\begin{array}{ccc}
\alpha=\left(\frac{b_0 l^2 s^2}{\left(b^2_0+l^2\right)^{5/2}}-1\right)^2&&
\beta=\frac{e^{\frac{2 b_0}{\sqrt{b_0^2+l^2}}} \left(b_0^2+l (l-s)\right) \left(b_0^2+l (l+s)\right)}{\left(b_0^2+l^2\right)^2}\\\\
\delta=\frac{l s \left(b_0-\sqrt{b_0^2+l^2}\right)}{\left(b_0^2+l^2\right)^{3/2} \sqrt{e^{-\frac{2 b_0}{\sqrt{b_0^2+l^2}}} \left(b_0^2+l^2\right)}}&&
\sigma=\frac{b_0^2 l^2 s^2}{\left(b_0^2+l^2\right)^4}-\frac{1}{b_0^2+l^2}\\\\
\rho=\left(m-\frac{b_0 l^2 m s^2}{\left(b_0^2+l^2\right)^{5/2}}\right)^2&&
\end{array}
\end{equation}

We consider $m=1$.

```python
alpha = ((b*l**2*s**2)/(b**2+l**2)**(5/2)-1)**2
```

```python
beta = (sp.exp((2*b)/sp.sqrt(b**2+l**2))*(b**2+l*(l-s))*(b**2+l*(l+s)))/(b**2+l**2)**2
```

```python
delta = (l*(b-sp.sqrt(b**2+l**2))*s)/((b**2+l**2)**(3/2)*sp.sqrt(sp.exp(-2*b/sp.sqrt(b**2+l**2))*(b**2+l**2)))
```

```python
sigma = (-1)/(b**2+l**2)+(b**2*l**2*s**2)/(b**2+l**2)**4
```

```python
rho = (1-(b*l**2*s**2)/(b**2+l**2)**(5/2))**2
```

# The effective potential 

The effective potential is given by 

\begin{equation}
V=\sqrt{\frac{\rho -L^2 \sigma }{\beta }+\left(\frac{\delta  L}{\beta }\right)^2}-\frac{\delta  L}{\beta }
\end{equation}

```python
Veff = -(delta*L/beta)+sp.sqrt((delta*L/beta)**2+(rho-sigma*L**2)/beta)
```

\begin{equation}
\begin{aligned}
\mathcal{V}_{eff}(l,b_0,s,\mathcal{L})&=\frac{l s \mathcal{L} \left(\sqrt{b_0^2+l^2}-b_0\right) \sqrt{e^{-\frac{2 b_0}{\sqrt{b_0^2+l^2}}} \left(b_0^2+l^2\right)}}{\sqrt{b_0^2+l^2} \left(\left(b_0^2+l^2\right)^2-l^2 s^2\right)}\\
&+\sqrt{\frac{e^{-\frac{2 b_0}{\sqrt{b_0^2+l^2}}} \left(\left(b_0^2+l^2\right)^2 \left(b_0^2+l (l-s)\right) \left(b_0^2+l (l+s)\right) \left(\frac{\mathcal{L}^2 \left(\left(b_0^2+l^2\right)^3-b_0^2 l^2 s^2\right)}{\left(b_0^2+l^2\right)^4}+\left(1-\frac{b_0 l^2 s^2}{\left(b_0^2+l^2\right)^{5/2}}\right)^2\right)+l^2 s^2 \mathcal{L}^2 \left(b_0-\sqrt{b_0^2+l^2}\right)^2\right)}{\left(\left(b_0^2+l^2\right)^2-l^2 s^2\right)^2}}
\end{aligned}
\end{equation}

which can be expressed as

\begin{equation}
\begin{aligned}
\mathcal{V}_{eff}(\tilde{l},\tilde{s},\tilde{\mathcal{L}})&=\frac{\tilde{l} \tilde{s} \tilde{\mathcal{L}} \left(\sqrt{1+\tilde{l}^2}-1\right) \sqrt{e^{-\frac{2}{\sqrt{1+\tilde{l}^2}}} \left(1+\tilde{l}^2\right)}}{\sqrt{1+\tilde{l}^2} \left(\left(1+\tilde{l}^2\right)^2-\tilde{l}^2 \tilde{s}^2\right)}\\
%
&+\sqrt{\frac{e^{-\frac{2}{\sqrt{1+\tilde{l}^2}}} \left(\left(1+\tilde{l}^2\right)^2 \left(1+\tilde{l}(\tilde{l}-\tilde{s})\right) \left(1+\tilde{l}(\tilde{l}+\tilde{s})\right) \left(\frac{\tilde{\mathcal{L}}^2 \left(\left(1+\tilde{l}^2\right)^3-\tilde{l}^2 \tilde{s}^2\right)}{\left(1+\tilde{l}^2\right)^4}+\left(1-\frac{\tilde{l}^2 \tilde{s}^2}{\left(1+\tilde{l}^2\right)^{5/2}}\right)^2\right)+\tilde{l}^2 \tilde{s}^2 \tilde{\mathcal{L}}^2 \left(1-\sqrt{1+\tilde{l}^2}\right)^2\right)}{\left(\left(1+\tilde{l}^2\right)^2-\tilde{l}^2 \tilde{s}^2\right)^2}}.
\end{aligned}
\end{equation}

Where
\begin{equation}
\begin{array}{ccc}
\tilde{l}=\frac{l}{b_0}&\text{The proper length}.& \text{Adimensional}\\
\tilde{s}=\frac{s}{b_0}=\frac{S}{b_0m}&\text{The spin of the particle.}&\text{Adimensional}\\
\tilde{\mathcal{L}}=\frac{\mathcal{L}}{b_0}=\frac{L}{b_0m}&\text{The angular momentum}&\text{Adimensional}.
\end{array}
\end{equation}

Here $b_0$ is the throat of the wormhole, and it is interpreted as the mass of the wormhole in the Newtonian limit, i.e. $b_0=M$ (see PRD 90, 024071 (2014)).

The symmetries in $\mathcal{V}_{eff}$ can be obtained from the first term, the one involving $ls\mathcal{L}$.

|  $\tilde{l}$ |  $\tilde{s}$ | $\tilde{\mathcal{L}}$|  Configuration|
|------|------|--------------|---------------|
|  +   |   +  |      +       |      Plus     |
|  +   |   +  |      -       |      Minus    |
|  +   |   -  |      +       |      Minus    |
|  +   |   -  |      -       |      Plus     |
|  -   |   -  |      -       |      Minus    |
|  -   |   -  |      +       |      Plus     |
|  -   |   +  |      -       |      Plus     |
|  -   |   +  |      +       |      Minus    |

From which we have the following symmetries 

\begin{equation}
\begin{aligned}
\mathcal{V}_{eff}(\tilde{l},\tilde{s},\tilde{\mathcal{L}})&=\mathcal{V}_{eff}(\tilde{l},-\tilde{s},-\tilde{\mathcal{L}})=\mathcal{V}_{eff}(-\tilde{l},\tilde{s},-\tilde{\mathcal{L}})=\mathcal{V}_{eff}(-\tilde{l},-\tilde{s},\tilde{\mathcal{L}})\\\\
\mathcal{V}_{eff}(\tilde{l},\tilde{s},-\tilde{\mathcal{L}})&=\mathcal{V}_{eff}(\tilde{l},-\tilde{s},\tilde{\mathcal{L}})=\mathcal{V}_{eff}(-\tilde{l},-\tilde{s},-\tilde{\mathcal{L}})=\mathcal{V}_{eff}(-\tilde{l},\tilde{s},\tilde{\mathcal{L}})
\end{aligned}
\end{equation}

```python
def V(l,b,s,L):
    return -L*l*s*(b - np.sqrt(b**2 + l**2))*(b**2 + l**2)**0.5*np.exp(-2*b/np.sqrt(b**2 + l**2))/(np.sqrt((b**2 + l**2)*np.exp(-2*b/np.sqrt(b**2 + l**2)))*(b**2 + l*(l - s))*(b**2 + l*(l + s))) + np.sqrt(L**2*l**2*s**2*(b - np.sqrt(b**2 + l**2))**2*np.exp(-2*b/np.sqrt(b**2 + l**2))/((b**2 + l*(l - s))**2*(b**2 + l*(l + s))**2) + (b**2 + l**2)**2*(-L**2*(b**2*l**2*s**2/(b**2 + l**2)**4 - 1/(b**2 + l**2)) + (-b*l**2*s**2*(b**2 + l**2)**(-2.5) + 1)**2)*np.exp(-2*b/np.sqrt(b**2 + l**2))/((b**2 + l*(l - s))*(b**2 + l*(l + s))))
    
    
```

# Superluminal condition 

There is one important feature which appears due to the spin of the particle that
should be considered. That is the superluminal bound on the particles motion. As it was mentioned before, the dynamical four-momentum and kinematical four-velocity of the spinning particle are not always parallel. Therefore, the normalization $u_\alpha u^\alpha = 1$ does not hold while $p_\alpha p^\alpha = m^2$ is satisfied. As the spinning particle approaches the center of the spacetime, its four-velocity increases and eventually, for certain values of the spin and radius, some components of the four-velocity may diverge as $u_\alpha u^\alpha\rightarrow \infty$. Before this to happen, the particles motion crosses the boundary between timelike and spacelike trajectories. Of course, spacelike (i.e., superluminal) motion is physically meaningless, and the transition to $u_\alpha u^\alpha > 0$ is not allowed for real particles. Therefore, one must impose a further bound, called superluminal bound, which is defined by the relation $u_\alpha u^\alpha = 0$. Hence, for the particle to move always in the timelike region, it is necessaryto impose the following condition [PRD100, 104052 (2019)]

\begin{equation}
\frac{u_\alpha u^\alpha}{(u^t)^2}=g_{tt}+g_{ll}(u^l)^2+g_{\phi\varphi}(u^\varphi)^2\leq 0,
\end{equation}

with the equality holding for the superluminal bound. To compute $u^l$ and $u^\varphi$, we use a method developed in Ref.[S. A. Hojman and F. A. Asenjo, Classical Quantum Gravity30, 025008 (2013)]. This method uses the MPD equations

\begin{equation}
\label{1}
\begin{aligned}
\frac{DS^{\alpha\beta}}{d\lambda}&=p^\alpha u^\beta-u^\alpha p^\beta\\
%
\frac{Dp^\alpha}{d\lambda}&=-\frac{1}{2}R^{\alpha}_{\;\beta\delta\sigma}u^\beta 
S^{\delta\sigma}
\end{aligned}
\end{equation}

where $D/d\lambda\equiv u^\alpha \nabla_\alpha$ is the projection of the covariant derivative along the particle's trajectory, $\lambda$ is the affine parameter, $u^\alpha$ is the Kinematical four-velocity, $p^\alpha$ is the four-momentum of the particle, and $R^{\alpha}_{\;\beta\delta\sigma}$ the Riemann tensor.From the first MPD equation, we obtain the following system of equations for the non-zero components of $S^{\alpha\beta}$,

\begin{equation}
\begin{aligned}
\frac{DS^{tl}}{d\lambda}&=p^tu^l-u^tp^l\\
\frac{DS^{t\varphi}}{d\lambda}&=p^tu^\varphi-u^tp^\varphi\\
\frac{DS^{l\varphi}}{d\lambda}&=p^lu^\varphi-u^lp^\varphi.
\end{aligned}
\end{equation}

We can express the last system of equations in terms of $S^{\varphi l}$. To do so, let's consider the firts equation. From the Tulczyjew-SSC condition $S^{\alpha\beta}p_\alpha=0$, we obtain

\begin{equation}
S^{tl}p_t+S^{\varphi l}p_\varphi=0.
\end{equation}

Applying the operator $D/d\lambda$, we get

\begin{equation}
\frac{D}{d\lambda}(S^{tl}p_l)+\frac{D}{d\lambda}(S^{\varphi l}p_\varphi)=
p_t\frac{D S^{tl}}{d\lambda}+S^{tl}\frac{D p_t}{d\lambda}+p_\varphi\frac{D S^{\varphi l}}{d\lambda}+S^{\varphi l}\frac{D p_\varphi}{d\lambda}=0,
\end{equation}

from which, after recalling that $S^{tl}=-\frac{p_\varphi}{p_t}S^{\varphi l}$, solving for $\frac{DS^{tl}}{d\lambda}$, and replacing in the MPD equation, we obtain

\begin{equation}
\frac{DS^{tl}}{d\lambda}=\frac{p_\varphi S^{\varphi l}}{(p_t)^2}\frac{Dp_t}{d\lambda}
-\frac{p_\varphi}{p_t}\frac{DS^{\varphi l}}{d\lambda}-\frac{S^{\varphi l}}{p_t}\frac{Dp_\varphi}{d\lambda}=p^tu^l-u^tp^l.
\end{equation}

In a similar way, we proceed with $\frac{D S^{t\varphi}}{d\lambda}$. From the Tulczyjew-SSC condition

\begin{equation}
S^{\alpha\varphi}p_\alpha=S^{l\varphi}p_l+S^{t\varphi}p_t=0.
\end{equation}

Then, applaying the operator  $D/d\lambda$, we obtain

\begin{equation}
\frac{D}{d\lambda}(S^{l\varphi}p_l)+\frac{D}{d\lambda}(S^{t\varphi}p_t)=
p_l\frac{D S^{l\varphi}}{d\lambda}+S^{l\varphi}\frac{Dp_l}{d\lambda}
+p_t\frac{DS^{t\varphi}}{d\lambda}+S^{t\varphi}\frac{Dp_t}{d\lambda}=0.
\end{equation}

Recalling that $S^{t\varphi}=\frac{p_l}{p_t}S^{\varphi l}$, taking into account the $S^{l\varphi}=-S^{\varphi l}$, solving for $\frac{D S^{t\varphi}}{d\lambda}$, and replacing into the MPD equatio, we obtain 

\begin{equation}
\frac{DS^{t\varphi}}{d\lambda}=\frac{p_l}{p_t}\frac{DS^{\varphi l}}{d\lambda}+\frac{S^{\varphi l}}{p_t}\frac{Dp_l}{d\lambda}-\frac{p_lS^{\varphi l}}{(p_t)^2}\frac{Dp_t}{d\lambda}=p^tu^\varphi-u^t p^\varphi.
\end{equation}

Now, from the MPD equation for $S^{\varphi l}$, we obtain

\begin{equation}
\begin{aligned}
\frac{D S^{tl}}{d\lambda}&=\frac{S^{\varphi l}}{p_t}\left(\frac{p_\varphi}{p_t}\frac{Dp_t}{d\lambda}-\frac{Dp_\varphi}{d\lambda}\right)-\frac{p_\varphi}{p_t}(p^\varphi u^l-u^\varphi p^l)=p^tu^l-u^tp^l\\
%
\frac{D S^{t\varphi}}{d\lambda}&=\frac{S^{\varphi l}}{p_t}\left(\frac{D p_l}{d\lambda}-\frac{p_l}{p_t}\frac{D p_t}{d\lambda}\right)+\frac{p_l}{p_t}(p^\varphi u^l-u^\varphi p^l)=p^tu^\varphi-u^tp^\varphi.
\end{aligned}
\end{equation}

Then, after multiplying by $p_t$ and organizing, the last system reduces to 

\begin{equation}
\begin{aligned}
\frac{S^{\varphi l}}{p_t}\left(p_\varphi\frac{Dp_t}{d\lambda}-p_t\frac{Dp_\varphi}{d\lambda}\right)&=(p_\varphi p^\varphi+p_tp^t)u^l-p^lp_\varphi u^\varphi-p^lp_tu^t\\
%
\frac{S^{\varphi l}}{p_t}\left(p_t\frac{Dp_l}{d\lambda}-p_l\frac{Dp_t}{d\lambda}\right)&=(p_lp^l+p_tp^t)u^\varphi-p^\varphi p_lu^l-p^\varphi p_tu^t.
\end{aligned}
\end{equation}

Now, we use the second MPD equation to obtain $\frac{Dp_t}{d\lambda}$, $\frac{Dp_\varphi}{d\lambda}$, and $\frac{Dp_l}{d\lambda}$. Hence from 

\begin{equation}
\frac{Dp_\alpha}{d\lambda}=-\frac{1}{2}R_{\alpha\beta\delta\sigma}u^\beta S^{\delta\sigma}
\end{equation}

we obtain

\begin{equation}
\frac{Dp_t}{d\lambda}=-\frac{1}{2}\left(2R_{tt\delta\sigma}u^tS^{\delta\sigma}
+2R_{tl\delta\sigma}u^lS^{\delta\sigma}+2R_{t\varphi \delta\sigma}u^\varphi S^{\delta\sigma}\right).
\end{equation}

The 2 factor in the Riemann tensor comes due to the antisymmetry of both $S^{\alpha\beta}$ and $R_{\alpha\beta\delta\sigma}$. Therfore, we have to count twice in the sum because $R_{ab\delta\sigma}S^{\delta\sigma}=R_{ab\sigma\delta}S^{\sigma\delta}$. Thus, we get

\begin{equation}
\frac{Dp_t}{d\lambda}=-R_{tltl}u^l S^{tl}-R_{t\varphi t\varphi}u^\varphi S^{t\varphi}
\end{equation}

Recalling that $S^{tl}=-\frac{p_\varphi}{p_t}S^{\varphi l}$ and $S^{t\varphi}=\frac{p_l}{p_t}S^{\varphi l}$, the last equation reduces to

\begin{equation}
\frac{Dp_t}{d\lambda}=\frac{S^{\phi l}}{p_t}\left(p_\varphi R_{tltl}u^l-p_l R_{t\varphi t\varphi}u^\varphi\right).
\end{equation}

Following a similar procedure, we obtain the followin relations for $\frac{Dp_l}{d\lambda}$, and $\frac{Dp_\varphi}{d\lambda}$ 

\begin{equation}
\begin{aligned}
\frac{Dp_l}{d\lambda}&=\frac{S^{\varphi l}}{p_t}(p_\varphi R_{lttl}u^t+p_tR_{l\varphi l \varphi}u^\varphi)\\
%
\frac{Dp_\varphi}{d\lambda}&=\frac{S^{\varphi l}}{p_t}(p_tR_{\varphi ll\varphi}u^l-p_lR_{\varphi tt\varphi}u^t)
\end{aligned}
\end{equation}

Finally, replacing into the systme of equations, we obtain

\begin{equation}
\begin{aligned}
u^l\left\{(p_\varphi)^2\left[g^{\varphi\varphi}+R_{tllt}\left(\frac{S^{\varphi l}}{p_t}\right)^2\right]+(p_t)^2\left[g^{tt}+R_{\varphi ll\varphi}\left(\frac{S^{\varphi l}}{p_t}\right)^2\right]\right\}&=u^\varphi\left\{g^{ll}+R_{\varphi tt\varphi}\left(\frac{S^{\varphi l}}{p_t}\right)^2\right\}p_lp_\varphi\\
&+u^t\left\{g^{ll}+R_{\varphi tt \varphi}\left(\frac{S^{\varphi l}}{p_t}\right)^2\right\}p_lp_t\\\\
%
u^\varphi\left\{(p_l)^2\left[g^{ll}+\left(\frac{S^{\varphi l}}{p_t}\right)^2R_{\varphi tt\varphi}\right]+(p_t)^2\left[g^{tt}+\left(\frac{S^{\varphi l}}{p_t}\right)^2R_{\varphi ll \varphi}\right]\right\}&=u^l\left\{g^{\varphi\varphi}+R_{lttl}\left(\frac{S^{\varphi l}}{p_t}\right)^2\right\}p_lp_\varphi\\
&+u^t\left\{g^{\varphi\varphi}+R_{lttl}\left(\frac{S^{\varphi l}}{p_t}\right)^2\right\}p_{\varphi} p_t
\end{aligned}
\end{equation}

Finally, solving the system we obtain 

\begin{equation}
\begin{array}{ccc}
u^l=\frac{\tilde{\mathcal{C}}}{\tilde{\mathcal{B}}}\frac{p_l}{p_t}&\text{and}&
u^\varphi=\frac{\tilde{\mathcal{A}}}{\tilde{\mathcal{B}}}\frac{p_\varphi}{p_t}
\end{array}
\end{equation}

where we have defined

\begin{equation}
\begin{array}{ccc}
\tilde{\mathcal{A}}=g^{\varphi\varphi}+R_{tllt}\left(\frac{S^{\varphi l}}{p_t}\right)^2,&
\tilde{\mathcal{B}}=g^{tt}+R_{\varphi ll\varphi}\left(\frac{S^{\varphi l}}{p_t}\right)^2,&
\tilde{\mathcal{C}}=g^{ll}+\left(\frac{S^{\varphi l}}{p_t}\right)^2R_{\varphi tt\varphi}.
\end{array}
\end{equation}

It is important to point out that following the gauge choices and invariant relations in Ref.[S. A. Hojman and F. A. Asenjo, Classical Quantum Gravity30, 025008 (2013)], the notation $\lambda = t$ is adopted.

Now, replacing into the superluminal bound condition

\begin{equation}
g_{tt}+g_{ll}(u^l)^2+g_{\varphi\varphi}(u^{\varphi})^2\leq 0
\end{equation}

we have
\begin{equation}
g_{tt}(\tilde{\mathcal{B}})^2(p_t)^2+g_{ll}(\tilde{\mathcal{C}})^2(p_l)^2+g_{\varphi\varphi}(\tilde{A})^2(p_\varphi)^2\leq 0,
\end{equation}

and from the conservation of the four momentum $p_\alpha p^\alpha=-m^2$, 

\begin{equation}
g^{ll}(p_l)^2=-g^{tt}(p_t)^2-g_{\varphi\varphi}(p_\varphi)^2-m^2.
\end{equation}

Hence, the superluminal bound condition reduces to

\begin{equation}
\left(\frac{p_t}{m}\right)^2\left\{g_{tt}(\tilde{\mathcal{B}})^2-\frac{(\tilde{\mathcal{C}})^2}{g_{tt}}\right\}+\left(\frac{p_\varphi}{m}\right)^2\left\{
g_{\varphi\varphi}(\tilde{\mathcal{A}})^2-\frac{(\tilde{\mathcal{C}})^2}{g_{\varphi\varphi}}\right\}-g_{ll}(\tilde{\mathcal{C}})^2\leq 0
\end{equation}

with

\begin{equation}
\begin{aligned}
\mathcal{X}=g_{tt}(\tilde{\mathcal{B}})^2-\frac{(\tilde{\mathcal{C}})^2}{g_{tt}}&=e^{\frac{2}{\sqrt{1+\tilde{l}^2}}}\left\{\left(\frac{\tilde{l}^2\tilde{s}^2}{(1+\tilde{l}^2)^\frac{5}{2}}-1\right)^2-\frac{[(1+\tilde{l}^2)^2-\tilde{s}^2]^2}{(1+\tilde{l}^2)^4}\right\}\\
\mathcal{Y}=g_{\varphi\varphi}(\tilde{\mathcal{A}})^2-\frac{(\tilde{\mathcal{C})^2}}{g_{\varphi\varphi}}&=\frac{1}{(1+\tilde{l}^2)^8}\left\{\left[(1+\tilde{l}^2)^\frac{7}{2}-(1-\tilde{l}^2(1+2\tilde{l}^2)+\tilde{l}^2\sqrt{1+\tilde{l}^2})s^2\right]^2-(1+\tilde{l}^2)^2
\left[(1+\tilde{l}^2)^\frac{5}{2}-\tilde{l}^2\tilde{s}^2\right]\right\}\\
\mathcal{Z}=g_{ll}(\tilde{\mathcal{C}})^2&=1-\frac{\tilde{l}^2\tilde{s}^2}{(1+\tilde{l}^2)^\frac{5}{2}}\\
\end{aligned}
\end{equation}

and

\begin{equation}
\begin{aligned}
\frac{p^t}{m}&=-\frac{\frac{\tilde{l} \sqrt{e^{-\frac{2}{\sqrt{\tilde{l}^2+1}}} \left(\tilde{l}^2+1\right)} \tilde{s} \tilde{\mathcal{L}}}{\left(\tilde{l}^2+1\right)^{5/2}}+\mathcal{E}}{1-\frac{\tilde{l}^2 s^2}{\left(\tilde{l}^2+1\right)^{5/2}}}\\
\frac{p^\varphi}{m}&=\frac{\frac{\tilde{l} \tilde{s} \mathcal{E}}{\sqrt{e^{-\frac{2}{\sqrt{\tilde{l}^2+1}}} \left(\tilde{l}^2+1\right)}}+\tilde{\mathcal{L}}}{1-\frac{\tilde{l}^2 \tilde{s}^2}{\left(\tilde{l}^2+1\right)^{5/2}}}
\end{aligned}
\end{equation}


# Definitions

```python
def XX(l,s):
    return np.exp(2/np.sqrt(1+l**2))*((l**2*s**2/(l**2+1)**(5/2)-1)**2-((1+l**2)**2-s**2)**2/(1+l**2)**4)
```

```python
def YY(l,s):
    return (-(1+l**2)**2*((1+l**2)**(5/2)-l**2*s**2)**2 + ((1+l**2)**(7/2) - (1-l**2-2*l**4+l**2*np.sqrt(1+l**2))*s**2)**2)/(1+l**2)**8
```

```python
def ZZ(l,s):
    return 1-(l**2*s**2)/(1+l**2)**(5/2)
```

```python
def pt(l,s,LL,EE):
    return -(EE+(l*s*LL*np.sqrt(np.exp(-2/np.sqrt(1+l**2))*(1+l**2)))/(1+l**2)**(5/2))/(1-l**2*s**2/(1+l**2)**(5/2))
```

```python
def pphi(l,s,LL,EE):
    return (LL+(l*s*EE)/(np.sqrt(np.exp(-2/np.sqrt(1+l**2))*(1+l**2))))/(1-l**2*s**2/(1+l**2)**(5/2))
```
